```python
import pandas as pd
import numpy as np
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia= SentimentIntensityAnalyzer()
```

```python
reviews = pd.read_csv("game_reviews.csv", index_col= False)
```

```python
reviews["review"]=reviews["review"].str.replace(r"[\"\'\|\?\=\.\@\#\*\,\.\!\/\:\-\‚ô•\<\>√â\+\D]", '',regex=True)
```

```python
#from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Verificar que 'review' sea una cadena de texto antes de aplicar la función
def get_sentiment_scores(x):
    if isinstance(x, str):
        return sia.polarity_scores(x)
    else:
        return None

# Aplicar la función al DataFrame
reviews['sentiment_scores'] = reviews['review'].apply(get_sentiment_scores)
```

```python
reviews['sentiment_scores'] = reviews['review'].apply(lambda x: sia.polarity_scores(x))
```

```python
def get_sentiment_score(text):
    if text == '':
        return 1
    sentiment = sia.polarity_scores(text)
    compound_score = sentiment['compound']
    if compound_score >= 0.05:
        return 2
    elif compound_score <= 0.05:
        return 0
    else:
        return 1
```

```python
reviews['sentiment_analysis'] = reviews ['review'].apply(get_sentiment_score)
```

```python
reviews.drop(columns=['review'], inplace= True) 
```

```python
reviews["Sentiment"]=reviews["review"].astype(str).apply(lambda i: sia.polarity_scores(i)["compound"])
```

```python
reviews["review"].replace("", np.nan, inplace=True)
```

```python
Sentiment_Result=[]

for i in range(len(reviews)):
    if reviews["Sentiment"].iloc[i]==0:
        Sentiment_Result.append(1)
    elif reviews["Sentiment"].iloc[i]<0:
        Sentiment_Result.append(0)
    elif reviews["Sentiment"].iloc[i]>0:
        Sentiment_Result.append(2)
```

```python
reviews["Sentiment_Result"]=pd.DataFrame(data=Sentiment_Result, index=None)
```

```python
reviews.drop(columns=["last_edited", "funny"], inplace=True)
```
