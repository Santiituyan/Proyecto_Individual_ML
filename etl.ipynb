```python
import pandas as pd
import numpy as np
import ast
```

```python
rows = []

with open ('australian_user_reviews.json', encoding="MacRoman") as f:
    for line in f.readlines():
        rows.append(ast.literal_eval(line))

df_reviews = pd.DataFrame(rows)
```

```python
df_games = pd.read_json('output_steam_games.json',lines=True)
```

```python
rows = []

with open ('australian_users_items.json' ,encoding='MacRoman') as f:
    for line in f.readlines():
      rows.append(ast.literal_eval(line))
        

df_items = pd.DataFrame(rows)
```

```python
df_games.dropna(thresh=5, inplace=True)
```

```python
df_reviews.dropna(thresh=5, inplace=True)
```

```python
df_items.dropna(thresh=5, inplace=True)
```

```python
data_desanidada = []

for index, row in df_items.iterrows():
    user_id = row['user_id']
    items_count = row['items_count']
    steam_id = row['steam_id']
    user_url = row['user_url']
    items = row['items']
    
    for i in items:   
        new_row = {
        'user_id': user_id,
        'items_count': items_count,
        'steam_id' : steam_id,
        'user_url' : user_url,
        'item_id': i.get('item_id', ''),
        'item_name': i.get('item_name', ''),
        'playtime_forever': i.get('playtime_forever', ''),
        'playtime_2weeks': i.get('playtime_2weeks', '')
        }
        
        data_desanidada.append(new_row)

df_i_desanidado = pd.DataFrame(data_desanidada)
```

```python
df_i_desanidado.to_csv("game_items.csv", index=False)
```
